```python
def valida_nome(nome:str) -> bool:
    ''' Função que verifica a entrada (string) do dado caso seja diferente de alfabético
  e espaço. '''
  
    for caracter in nome:
        if caracter.isalpha() or caracter.isspace():
            valida_nome = True
        else:
            valida_nome = False
            break
    return valida_nome
```

```python
def corrige_nome(nome:str) -> str:
    ''' Função que corrige a entrada (string) do dado caso seja diferente de alfabético
  e espaço. '''

    nome_validacao = valida_nome(nome)
    
    if nome_validacao:
        nome_corrigido = nome
    else:
        nome_corrigido = [caracter for caracter in nome if caracter.isalpha() or caracter.isspace()]
        nome_corrigido = ''.join(nome_corrigido)
    return nome_corrigido
```

```python
def valida_email(email:str) -> bool:
    ''' Função que verifica a entrada (string) do dado caso seja diferente de alfabético
  e espaço. '''
  
    for caracter in email:
        if caracter.isalpha() or caracter.isnumeric() or caracter in '@._':
            valida_email = True
        else:
            valida_email = False
            break
```

```python
def corrige_email(email:str) -> str:
    ''' Função que corrige a entrada (string) do dado caso seja diferente de alfabético
  e espaço. '''

    email_validacao = valida_email(email)

    if email_validacao:
        email_validacao = email
    else:
        email_validacao = [caracter for caracter in email if (caracter.isalpha() or caracter.isnumeric() or caracter in '@._')]
        email_validacao = retira_arroba(email_validacao)
        
    return email_validacao
```

```python
def retira_arroba(email:str) -> str:
    '''Função que garante a existência de somente um arroba na função'''
    
    arroba_corrigido = list(email)
    contador = 0
    for index in range(len(email)-1 , 0, -1):
        if arroba_corrigido[index] == '@':
            contador += 1
            
            if contador > 1:
                arroba_corrigido.remove(arroba_corrigido[index]) 
    return ''.join(arroba_corrigido)
```

```python
def entradas_musico() -> list:
    '''Função que recebe todas as entradas do usuário e dá o devido tratamento.'''
    
    nome_musico = input('\nDigite o nome do músico a ser cadastrado: ').rstrip().lstrip().lower()

    nome_musico = corrige_nome(nome_musico)

    email_musico = input('Digite o e-mail do músico a ser cadastrado: ').rstrip().lstrip().lower()

    email_musico = corrige_email(email_musico)

    genero_musical = input('Digite o gênero musical a ser cadastrado (mínimo 1): ')
    
    genero_musical = padroniza_lista(genero_musical)
    
    instrumento_musical = input('Digite o instrumento musical a ser cadastrado (mínimo 1): ')
    
    instrumento_musical = padroniza_lista(instrumento_musical)

    return [[nome_musico, email_musico, genero_musical, instrumento_musical]]
```

```python
def padroniza_lista(lista: list) -> list:
    '''Função que padroniza as entradas, transformando string em lista'''
    lista_padronizada = [elemento.rstrip().lstrip().lower() for elemento in lista]
    juncao = ''.join(lista_padronizada)
    return juncao if len(lista_padronizada) == 1 else juncao.split(',')
```

```python
def leitura_arquivo():
    '''Função para leitura do arquivo csv'''
    try:
        with open('musicos.csv', 'r') as arquivo:    
            leitura = list(csv.reader(arquivo, delimiter = ';', lineterminator = '\n'))
        return leitura
    except:
            print('O arquivo csv precisa ser criado!')
```

```python
def escrita_arquivo(dados_musico, tipo) -> list :
    '''Função para escrita do arquivo csv'''
    with open('musicos.csv', tipo) as arquivo:
        armazenamento = csv.writer(arquivo, delimiter = ';', lineterminator = '\n')
        armazenamento.writerows(dados_musico)
    print(f'\nA implementação foi realizada!')
    
    return armazenamento
```

```python
def cadastrar_musico() -> None:
    '''Função que cadastra o músico, caso este não esteja cadastrado.'''
    
    dados_musico = entradas_musico()
    leitura = leitura_arquivo()
    
    for elemento in leitura:
        if dados_musico[0][1] not in elemento[1]:
            armazenamento = escrita_arquivo(dados_musico, 'a')
            break
        elif dados_musico[0][1] in elemento[1]:
            print('E-mail já cadastrado! O cadastro não pode ser concluído.')
            break
```

```python
def buscar_musico(**dados_musico):
    
    quantidade_parametros = len(dados_musico)
    
    leitura = leitura_arquivo()
    
    lista_parametros = []
    musicos_filtrados = []
    
    for parametro in dados_musico:
        if parametro in ['nome','email','genero','instrumento']:
            lista_parametros.append(parametro)

    if quantidade_parametros > 1:
        
        busca_e_ou = input('Você quer buscar com base nos parâmetros simultaneamente ou individualmente?\ndigite "e" ou "ou": ')
    
    for lista in leitura:
        if quantidade_parametros == 1:
        
            parametro01 = dados_musico[lista_parametros[0]]
            if parametro01 in lista:
                musicos_filtrados.append(lista)
                
        elif quantidade_parametros == 2:
            
            parametro01 = dados_musico[lista_parametros[0]]
            parametro02 = dados_musico[lista_parametros[1]]
            
            if busca_e_ou == 'e':
                if parametro01 in lista and parametro02 in lista:
                    musicos_filtrados.append(lista)
            elif busca_e_ou == 'ou':
                if parametro01 in lista or parametro02 in lista:
                    musicos_filtrados.append(lista)
        
        elif quantidade_parametros == 3:  
        
            parametro01 = dados_musico[lista_parametros[2]]
            parametro02 = dados_musico[lista_parametros[3]]
            parametro03 = dados_musico[lista_parametros[4]]
   
            if parametro01 in lista and parametro02 in lista and parametro03 in lista:
                musicos_filtrados.append(lista)
            elif parametro01 in lista or parametro02 in lista or parametro03 in lista:
                musicos_filtrados.append(lista)
                
        elif quantidade_parametros == 4: 
        
            parametro01 = dados_musico[lista_parametros[0]]
            parametro02 = dados_musico[lista_parametros[1]]
            parametro03 = dados_musico[lista_parametros[2]]
            parametro04 = dados_musico[lista_parametros[3]]
            
            if parametro01 in lista and parametro02 in lista and parametro03 in lista and parametro04 in lista:
                musicos_filtrados.append(lista)
            elif parametro01 in lista or parametro02 in lista or parametro03 in lista or parametro04 in lista:
                musicos_filtrados.append(lista)
                
    return musicos_filtrados
```

```python
def modificar_musico(email:str) -> str:
    
    leitura = leitura_arquivo()
    for musico in leitura:
        if email == musico[1]:
            
            modificador  = input('Você deseja adicionar ou remover um item? ')
            item_a_modificar = input(f'Você deseja {modificador} gênero musical ou instrumento? ')
            
            if modificador == 'adicionar':
                if item_a_modificar == ('genero musical' or 'gênero musical'):
                    musico[2] = ast.literal_eval(musico[2])
                    musico[2].append(input(f'Adicione o {item_a_modificar} do músico {musico[0].title()}: '))
                    
                elif item_a_modificar == 'instrumento':
                    musico [3] = ast.literal_eval(musico[3])
                    musico[3].append(input(f'Adicione o {item_a_modificar} do músico {musico[0].title()}: ')) 
                                           
            elif modificador == 'remover':
                if item_a_modificar == ('genero musical' or 'gênero musical'):
                    musico [2] = ast.literal_eval(musico[2])
                    musico[2].remove(input(f'Exclua o {item_a_modificar} do músico {musico[0].title()} '))
                    
                elif item_a_modificar == 'instrumento':
                    musico [3] = ast.literal_eval(musico[3])
                    musico[3].remove(input(f'Exclua o {item_a_modificar} do músico {musico[0].title()} '))  
    
    escrita_arquivo(leitura,'w')
```

```python
def pre_filtro(quantidade_musicos):
    
    genero_banda = input('Qual o gênero musical da sua banda? ')
    
    instrumentos = [input(f'Instrumento {index+1} que deseja: ') for index in range(quantidade_musicos)]
    
    musicos = leitura_arquivo()
    
    musico_filtrado = []
    lista_lixo = []
    for instrumento in instrumentos:
        lista_temporaria = []
        for musico in musicos:
            if genero_banda in musico[2] and instrumento in musico[3]:
                if musico[1] not in lista_lixo:
                    if instrumento not in musico_filtrado:
                        musico_filtrado.append(instrumento)
                        lista_temporaria.append(musico[1])
                        lista_lixo.append(musico[1])
                    
                    if musico[1] not in lista_temporaria:
                        lista_temporaria.append(musico[1])
                        lista_lixo.append(musico[1])
                
                    
        musico_filtrado.append(lista_temporaria)        
    return musico_filtrado
```

```python
def ordena_email_instrumento(quantidade_musicos):
    
    lista = pre_filtro(quantidade_musicos)
    
    instrumentos_ordenados = []
    email_por_instrumento = []
    
    for elemento in lista:
        if type(elemento) == list:
            email_por_instrumento.append(elemento)
        else:
            instrumentos_ordenados.append(elemento)
            
    return instrumentos_ordenados, email_por_instrumento
```

```python
def combinacoes_banda(n, lista, lista_aux = []):
    
    if len(lista_aux) == n:
        return [lista_aux]

    lista_combinacoes = []
    
    for linha in lista[0]:
        lista_copia = lista_aux[:]
        lista_copia.append(linha)
        lista_combinacoes += combinacoes_banda(n, lista[1:], lista_copia)
    return lista_combinacoes
```

```python
def montar_bandas(n):
    
    instrumentos, email_desordenado = ordena_email_instrumento(n)
    email = combinacoes_banda(n, email_desordenado)
    
    for lista in email:
        for index in range(len(instrumentos)):
            print(f'{instrumentos[index].title()}: {lista[index]}', end = ' ')
        print()
        print('---'*30)

    
```

```python
def sair():
    print('\nFim do  programa! Obrigado por usar')
```

```python
def menu(entrada):

    opcoes = {'1': cadastrar_musico, '2': buscar_musico, '3': modificar_musico,'4': montar_bandas, '0': sair}
    return opcoes[entrada]
```

```python
# encoding: utf-8
# encoding: iso-8859-1
# encoding: win-1252

import ast
import csv
from pathlib import Path

opcao = input('Entre com a opção desejada no menu: ')
funcao = menu(opcao)

if opcao == '1':
    
    resposta = funcao()
    
elif opcao == '2':
    print('''Entre abaixo com a busca no formato de dicionario como no exemplo: 
Exemplo: nome,email,genero,instrumento''')
    
    parametros_lista = input('Parâmetros[separe os parâmetros somente por vírgula]: ').split(',')
    parametros_dicionario = {}
    for parametro in parametros_lista:
        parametros_dicionario[parametro] = input(f'Digite o {parametro} de busca: ').lower()
    resposta = funcao(**parametros_dicionario)

elif opcao == '3':
        
    email = input('Entre com o email do músico que deseja buscar: ')
    email = corrige_email(email)
    resposta = funcao(email)
        
elif opcao == '4':
    n = int(input('Quantos membros a sua banda terá? '))
    resposta = funcao(n)
    
elif opcao == '0':
    
    resposta = funcao()

if opcao == '2'  or opcao == '3':
    for linha in resposta:
        print(f'\n{linha}')
```
